```python
import tkinter as tk
from tkinter import messagebox

def submit():
    weight_ = int(weight_entry.get())
    location_ = int(location_entry.get())
    if location_ > 2 or location_ < 1: 
        messagebox.showerror("Order","Invalid Location Number (Input 1 or 2)",)
    else:
        calculate_price()
        
        
def calculate_price():
    weight_ = int(weight_entry.get())
    location_ = int(location_entry.get())
    root = tk.Tk()
    root.title("Adekunle & Simi's Delivery Service")
    root.geometry('500x250')
    
    if location_entry == 2 and weight_ >=10:
        price = 5000
    

    elif location_entry == 2 and weight_ <10:
        price = 3500
        
        
    elif location_entry == 1 and weight_ >=10:
        price  =10000
        

    else:
        price= 5000
    if location_ == 1:
        label_calc = tk.Label(root, text=f"From the Input Location (Ibeju-Lekki) And Weight ({weight_}KG)\n\nPrice = N{price} ONLY.")
        label_calc.pack()
        return
        
    else:    
        label_calc = tk.Label(root, text=f"From the Input Location (Epe) And Weight ({weight_}KG)\n\nPrice = N{price} ONLY.")
        label_calc.pack()
        return

    
    root.mainloop()
root = tk.Tk()
root.title("Adekunle & Simi's Delivery Service")
root.geometry('500x250')



#location label
location_label = tk.Label(root, text="Please Input your location (Use Number):\n1) Ibeju-Lekki\n2) Epe")
location_label.pack()
location_entry =tk.Entry(root)
location_entry.pack()

weight_label = tk.Label(root, text = "How much weight is the package(In KG)")
weight_label.pack()
weight_entry =tk.Entry(root)
weight_entry.pack()
submit_button = tk.Button(root, text="Submit",command=submit)
submit_button.pack()

root.mainloop()



```
